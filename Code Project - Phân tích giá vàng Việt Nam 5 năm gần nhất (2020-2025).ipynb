```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Bước 1: Tải dữ liệu giá vàng 5 năm gần nhất với tần suất hàng ngày
gold = yf.download("GC=F", period="5y", interval="1d")  # Lấy dữ liệu theo ngày

# Kiểm tra dữ liệu có bị trống không
if gold.empty:
    print("Không có dữ liệu để lưu! Vui lòng kiểm tra lại mã tài sản hoặc kết nối mạng.")
else:
    # Bước 2: Lưu dữ liệu vào file Excel
    file_name = "gia_vang_5_nam_theo_ngay.xlsx"
    gold.to_excel(file_name)
    print(f"File Excel đã được tạo: {file_name}")

    # Bước 3: Vẽ biểu đồ giá vàng theo từng ngày
    plt.figure(figsize=(12, 6))
    plt.plot(gold.index, gold["Close"], linestyle="-", color='gold', label="Giá vàng đóng cửa")

    # Tùy chỉnh biểu đồ
    plt.xlabel("Năm")
    plt.ylabel("Giá vàng (USD)")
    plt.title("Biểu đồ giá vàng theo ngày trong 5 năm gần nhất")
    plt.legend()
    plt.grid()

    # Hiển thị biểu đồ
    plt.show()
```
