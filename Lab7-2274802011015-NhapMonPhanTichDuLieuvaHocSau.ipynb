```python
#CAU 4
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Tải dữ liệu cần thiết
try:
    nltk.download('punkt')
    nltk.download('stopwords')
except Exception as e:
    print(f"Lỗi khi tải dữ liệu NLTK: {e}")

def remove_stopwords(text, language='english'):
    try:
        stop_words = set(stopwords.words(language))
        word_tokens = word_tokenize(text)
        filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
        return filtered_text
    except Exception as e:
        print(f"Lỗi khi xử lý văn bản: {e}")
        return []

text = "This is an example sentence demonstrating the removal of stop words."
print("Văn bản sau khi loại bỏ stopwords:", remove_stopwords(text))
```

```python
#CAU 5
from nltk.corpus import stopwords

def ignore_some_stopwords(custom_keep_words, language='english'):
    all_stopwords = set(stopwords.words(language))
    filtered_stopwords = [word for word in all_stopwords if word not in custom_keep_words]
    return filtered_stopwords

print("Stopwords tiếng Anh sau khi bỏ qua 'not', 'no':",
      ignore_some_stopwords(['not', 'no'])[:10])
```

```python
#CAU 6
from nltk.corpus import wordnet
nltk.download('wordnet')

def wordnet_info(word):
    synsets = wordnet.synsets(word)
    if not synsets:
        return "Không tìm thấy từ trong WordNet"

    info = []
    for syn in synsets[:3]:  # Giới hạn 3 kết quả đầu
        info.append(f"Định nghĩa: {syn.definition()}")
        examples = syn.examples()
        if examples:
            info.append(f"Ví dụ: {examples[0]}")
    return "\n".join(info)

print("Thông tin về từ 'computer':")
print(wordnet_info('computer'))
```

```python
#CAU 7
from nltk.corpus import wordnet

def get_synonyms_antonyms(word):
    synonyms = set()
    antonyms = set()

    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

    return {"Từ đồng nghĩa": list(synonyms), "Từ trái nghĩa": list(antonyms)}

print("Từ đồng nghĩa và trái nghĩa của 'happy':")
print(get_synonyms_antonyms('happy'))
```

```python
#CAU 9
from nltk.corpus import wordnet

def noun_similarity(noun1, noun2):
    syn1 = wordnet.synsets(noun1, pos=wordnet.NOUN)
    syn2 = wordnet.synsets(noun2, pos=wordnet.NOUN)

    if not syn1 or not syn2:
        return 0.0

    return syn1[0].path_similarity(syn2[0])

print("Độ tương đồng giữa 'cat' và 'dog':", noun_similarity('cat', 'dog'))
```

```python
#CAU 10
from nltk.corpus import wordnet

def verb_similarity(verb1, verb2):
    syn1 = wordnet.synsets(verb1, pos=wordnet.VERB)
    syn2 = wordnet.synsets(verb2, pos=wordnet.VERB)

    if not syn1 or not syn2:
        return 0.0

    return syn1[0].path_similarity(syn2[0])

print("Độ tương đồng giữa 'run' và 'walk':", verb_similarity('run', 'walk'))
```

```python
#CAU 11
nltk.download('names')

from nltk.corpus import names

male_names = names.words('male.txt')
female_names = names.words('female.txt')

print(f"Số lượng tên nam: {len(male_names)}")
print(f"Số lượng tên nữ: {len(female_names)}")

print("\n10 tên nam đầu tiên:", male_names[:10])
print("10 tên nữ đầu tiên:", female_names[:10])
```

```python
#CAU 12
import random

def random_name_pairs(num_pairs=15):
    pairs = []
    for _ in range(num_pairs):
        male = random.choice(male_names)
        female = random.choice(female_names)
        pairs.append((male, female))
    return pairs

print("15 cặp tên ngẫu nhiên (nam, nữ):")
for i, pair in enumerate(random_name_pairs(), 1):
    print(f"{i}. {pair[0]} & {pair[1]}")
```

```python
#CAU 13
def last_letter_mapping():
    mapping = []
    for name in male_names:
        mapping.append((name[-1].lower(), 'male'))
    for name in female_names:
        mapping.append((name[-1].lower(), 'female'))
    return mapping

last_letter_map = last_letter_mapping()
print("\nMapping 10 mẫu đầu tiên:")
print(last_letter_map[:10])
```
